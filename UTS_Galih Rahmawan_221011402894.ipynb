```python
# pertanyaan 2 => Please print all Domain value in every company.


def get_company_domain():
    for company in company_name_list:
        company_name = company['name']
        company_details = company_detail_list.get(company_name)
        if company_details:
            domain = company_details['domain']
            num_clients = len(company_details['clients'])
            print(f"{company_name}: {domain}, relation: {num_clients} clients")
        else:
            print(f"{company_name}: Details not found")

get_company_domain()
```

```python
# pertanyaan 3 => Create a function that return employees with it's company domain.


def get_employees():
    employees_with_domain = []
    for employee in employee_name_list:
        employee_name = employee['name']
        company_name = employee_detail_list[employee_name]['company']
        company_domain = company_detail_list[company_name]['domain']
        full_name = employee_detail_list[employee_name]['full_name']
        employees_with_domain.append({
            "full_name": full_name,
            "company": company_name,
            "domain": company_domain
        })
    return employees_with_domain

print(json.dumps(get_employees(), indent=4))
```

```python
# pertanyaan 4 => Create a function that return companies with list of employees that are working in the company.


def get_employees():
    employees_by_company = {}
    for employee in employee_name_list:
        employee_name = employee['name']
        company_name = employee_detail_list[employee_name]['company']
        full_name = employee_detail_list[employee_name]['full_name']
        if company_name not in employees_by_company:
            employees_by_company[company_name] = []
        employees_by_company[company_name].append(full_name)
    
    result = []
    for company in company_name_list:
        company_name = company['name']
        employees = employees_by_company.get(company_name, [])
        result.append({
            "company": company_name,
            "employees": employees
        })
    
    return result

print(json.dumps(get_employees(), indent=4))
```
